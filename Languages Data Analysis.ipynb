```python
# relationship between number of classes and enrollment numbers for each language, also break down by semester
def findEnrollees(x, cdf):
    return cdf.loc[cdf['Course Title'] == x, 'Total'].sum()

S20 = df[(df['Year'] == 2020) & (df['Semester'] == 'Spring')]
points = S20['Course Title'].value_counts().rename_axis('Languages').reset_index(name='# Classes')
points['Enrollment Numbers'] = points.apply(lambda row: findEnrollees(row['Languages'], S20), axis=1)

for x in range(len(dfList)):
    temp = dfList[x]['Course Title'].value_counts().rename_axis('Languages').reset_index(name='# Classes')
    temp['Enrollment Numbers'] = temp.apply(lambda row: findEnrollees(row['Languages'], dfList[x]), axis=1)
    points = pd.concat([points, temp])
```

```python
# regression for above (show it's not very good)
dfTrain, dfTest = train_test_split(points, test_size=0.2, random_state=1000)
xTrain, xTest = dfTrain[['# Classes']], dfTest[['# Classes']]
yTrain, yTest = dfTrain[['Enrollment Numbers']], dfTest[['Enrollment Numbers']]
linreg = LinearRegression().fit(xTrain, yTrain)
print(f'Coefficient: {linreg.coef_[0][0]}')
print(f'Intercept: {linreg.intercept_[0]}')
```

```python
train_r2 = linreg.score(xTrain, yTrain)
test_r2 = linreg.score(xTest, yTest)
train_mse = mean_squared_error(linreg.predict(xTrain), yTrain)
test_mse = mean_squared_error(linreg.predict(xTest), yTest)
print(f'Training R2 Score: {train_r2:.4f}')
print(f'Test R2 Score: {test_r2:.4f}')
print(f'Training MSE: {train_mse:.2f}')
print(f'Test MSE: {test_mse:.2f}')
```

```python
train_res = linreg.predict(xTrain) - yTrain
test_res = linreg.predict(xTest) - yTest
fig, axs = plt.subplots(1, 2, figsize=(12,4))

axs[0].hist(train_res)
axs[0].set_title("Training Residuals Histogram")
axs[1].hist(test_res)
axs[1].set_title("Test Residuals Histogram");
```

```python
# how many classes and languages each prof teaches
profs = df[df['Instructor Full Name'] != 'Unknown']
dfListProfs = [profs[(profs['Year'] == 2020) & (profs['Semester'] == 'Fall')], profs[(profs['Year'] == 2021) & (profs['Semester'] == 'Spring')], 
    profs[(profs['Year'] == 2021) & (profs['Semester'] == 'Fall')], profs[(profs['Year'] == 2022) & (profs['Semester'] == 'Fall')], 
    profs[(profs['Year'] == 2022) & (profs['Semester'] == 'Spring')], profs[(profs['Year'] == 2023) & (profs['Semester'] == 'Fall')], 
    profs[(profs['Year'] == 2023) & (profs['Semester'] == 'Spring')], profs[(profs['Year'] == 2024) & (profs['Semester'] == 'Fall')], 
    profs[(profs['Year'] == 2024) & (profs['Semester'] == 'Spring')]]


def findLanguages(x, cdf):
    return len(cdf.loc[cdf['Instructor Full Name'] == x, 'Course Title'].unique())

S20 = profs[(profs['Year'] == 2020) & (profs['Semester'] == 'Spring')]
classesTeaching = S20['Instructor Full Name'].value_counts().rename_axis('Professor Name').reset_index(name='# Classes Teaching')
classesTeaching['# Languages Teaching'] = classesTeaching.apply(lambda row: findLanguages(row['Professor Name'], S20), axis=1)

for x in range(len(dfListProfs)):
    temp = dfListProfs[x]['Instructor Full Name'].value_counts().rename_axis('Professor Name').reset_index(name='# Classes Teaching')
    temp['# Languages Teaching'] = temp.apply(lambda row: findLanguages(row['Professor Name'], dfListProfs[x]), axis=1)
    classesTeaching = pd.concat([classesTeaching, temp])

classesTeaching = classesTeaching.sort_values(by=['# Languages Teaching', '# Classes Teaching'], ascending=[False, False])
classesTeaching.head(30)
```
